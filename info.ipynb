```
import pandas as pd
import pickle
import nltk
from nltk.corpus import stopwords
import os
import json
import re
import sys
import unicodedata
enero= pd.read_pickle("./drive/MyDrive/datasets/enero.pkl")
febrero= pd.read_pickle("./drive/MyDrive/datasets/febrero.pkl")
marzo= pd.read_pickle("./drive/MyDrive/datasets/marzo.pkl")
abril= pd.read_pickle("./drive/MyDrive/datasets/abril.pkl")
mayo= pd.read_pickle("./drive/MyDrive/datasets/mayo.pkl")
junio= pd.read_pickle("./drive/MyDrive/datasets/junio.pkl")
julio= pd.read_pickle("./drive/MyDrive/datasets/julio.pkl")
agosto= pd.read_pickle("./drive/MyDrive/datasets/agosto.pkl")
septiembre= pd.read_pickle("./drive/MyDrive/datasets/septiembre.pkl")
octubre= pd.read_pickle("./drive/MyDrive/datasets/octubre.pkl")
noviembre= pd.read_pickle("./drive/MyDrive/datasets/noviembre.pkl")
diciembre= pd.read_pickle("./drive/MyDrive/datasets/diciembre.pkl")
data=pd.concat([enero,febrero,marzo , abril, mayo,junio, julio, agosto,septiembre,octubre,noviembre,diciembre])
data['tweet']=data['tweet'].apply(lambda x: " ".join(x))
data=data[data['language']=='es']

def filter_stop_words(text):
    stop_words_list = stopwords.words('spanish')
    stop_words_list += stopwords.words('english')
    stop_words_list+=["arauz","correa","hoy","asÃ­","de","que","el","las","los","en","la","lo","hola","lunes","martes","jueves","viernes","sabado","domingo","miercoles","covid", "covid","covid19","coronavirus",'pues','mba']
    text_filtered = [word for word in text.split() if word.lower() not in stop_words_list]
    return text_filtered

#tweet=data[data['date']=='2020-04']

#data['tweet']=data['tweet'].apply(lambda x: " ".join(x))
tweet=data
tweet['tweet']=tweet['tweet'].apply(lambda x:filter_stop_words(x))
```

```
from sentiment_analysis_spanish import sentiment_analysis
count=0
def sentiment(tweet):
    sentiment = sentiment_analysis.SentimentAnalysisSpanish()
    analysis=sentiment.sentiment(tweet)
    global count
    print(f'....Analysis..........{count}/120000')
    count+=1
    if(analysis>=0.7):
        return 'POS'
    if(analysis<=0.3):
        return 'NEG'
    if(0.3< analysis <0.7):
        return 'NEU'
    
zw=tweet['sentiment']=tweet['tweet'].apply(lambda x : sentiment(" ".join(x)))
zw.to_pickle("./drive/MyDrive/datasets/analysis_sentiment.pkl")
```
